```python
# Interactive Graphs with Dash 
import dash_core_components as dcc
import dash_html_components as html 
from dash.dependencies import Output, Input
```

```python
# Data Exploration with Pandas (python)
```

```python
df = pd.read_csv("/content/Sales.csv") #Creating the dataframe "df" using the csv file
```

```python
print(df[:5]) #Print the first 5 rows
```

```python
print(df.iloc[:5, [2,3,5,10]]) #Print the first 5 rows and see the specific 2nd,3rd,5th and 10th column we want to see
```

```python
print(df.Genre.nunique()) #find out the unique categories under the "Genre" column
```

```python
print(df.Genre.unique()) #See the different categories under the "Genre" column
```

```python
print(sorted(df.Year.unique())) #Print and See the different years in order under the "Year" column
```

```python
# Data Visualization with Plotly (Python)
fig_pie = px.pie(data_frame=df, names='Genre', values='Japan Sales') #call a piechart, define the dataframe, names is the variable and values is the country under Sales column metrics
fig_pie = px.pie(data_frame=df, names='Genre', values='North American Sales') #call another piechart, define the dataframe, names is the variable and values is the country under Sales column metrics
fig_pie.show() #Show the Pie Chart
```

```python
fig_bar = px.bar(data_frame=df, x='Genre', y='World Sales')
fig_bar.show()
```

```python
fig_hist = px.histogram(data_frame=df, x='Year', y='World Sales')
fig_hist.show()
```

```python
app = dash.Dash(__name__)
```

```python
#Create a layout, within the layout, we need a Title, a dropdown, and a space for plotting the visual
app.layout=html.Div([
    html.H1("Graph Analysis with Charming Data"),
    dcc.Dropdown(id='genre-choice',
                 options=[{'label':x, 'value':x}
                          for x in sorted(df.Genre.unique())],
                 value='Action'
                 ),
    dcc.Graph(id='my-graph',
              figure={}),
])
```

```python
@app.callback(
    Output(component_id='my-graph', component_property='figure'),
    Input(component_id='genre-choice', component_property='value')
)
def interactive_graphs(value_genre):
    print(value_genre)
    dff = df[df.Genre==value_genre]
    fig = px.bar(data_frame=dff, x='Year', y='World Sales')
    return fig
```

```python
if __name__=='__main__':
    app.run_server()
```
